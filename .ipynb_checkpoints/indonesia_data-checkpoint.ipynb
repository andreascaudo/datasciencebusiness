Import Indonesia data

```python
import pandas as pd
import glob
import os


path = os.getcwd() + "/Data/Indonesia/indonesia-coronavirus-cases/cases.csv"

columns = ["total_cases","daily_cases","actual_cases","total_healed","daily_healed","total_dead","daily_dead",
                  "total_test","daily_test"]

dataset = pd.read_csv(path, index_col = 0)
dataset = dataset.fillna(0)
```

```python
dataset
```

Creating a new dataframe with our specific format

```python
indonesia_dataset = pd.DataFrame(index = dataset.index , columns = columns) 
indonesia_dataset = indonesia_dataset.fillna(0)
indonesia_dataset
```

Filling the new dataframe with the avaiable data

```python
from datetime import datetime, timedelta
day = []
for days in dataset.index:
    try:
        #from 2nd row ---- to subtract a day from the date we had to convert it to a date and then back to a string
        yesterday = (datetime.strptime(days,"%d-%b-%y")-timedelta(days=1)).strftime("%-d-%b-%y")
        indonesia_dataset.loc[days]["daily_test"] = dataset.loc[days]["acc_tested"] - dataset.loc[yesterday]["acc_tested"]
    except:
        #First row
        indonesia_dataset.loc[days]["daily_test"] = dataset.loc[days]["acc_tested"]
    finally:
        #Single row
        indonesia_dataset.loc[days]["total_cases"] = dataset.loc[days]["acc_confirmed"]
        indonesia_dataset.loc[days]["daily_cases"] = dataset.loc[days]["new_confirmed"]
        indonesia_dataset.loc[days]["actual_cases"] = dataset.loc[days]["isolated"]
        indonesia_dataset.loc[days]["total_healed"] = dataset.loc[days]["acc_released"]
        indonesia_dataset.loc[days]["daily_healed"] = dataset.loc[days]["new_released"]
        indonesia_dataset.loc[days]["total_dead"] = dataset.loc[days]["acc_deceased"]
        indonesia_dataset.loc[days]["daily_dead"] = dataset.loc[days]["new_deceased"]
        indonesia_dataset.loc[days]["total_test"] = dataset.loc[days]["acc_tested"]
        

indonesia_dataset
```
