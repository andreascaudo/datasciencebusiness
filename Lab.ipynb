Creare un array di Datafreme per regioni italiane conteneti le informazioni giornaliere.


```python
import pandas as pd
import glob


files = ["dpc-covid19-ita-regioni-20200224.csv"]

path = os.getcwd() + "/Data/Italia/COVID-19-master/dati-regioni/"
all_files = glob.glob(path + "/*.csv")

all_files.remove(os.getcwd() + "/Data/Italia/COVID-19-master/dati-regioni/" + "dpc-covid19-ita-regioni.csv")
all_files.remove(os.getcwd() + "/Data/Italia/COVID-19-master/dati-regioni/" + "dpc-covid19-ita-regioni-latest.csv")
            
regions = {} 

for f in all_files:
    dataset = pd.read_csv(f, index_col = 0)
    for reg in dataset["denominazione_regione"].values:
        try:
            regions[reg] = regions[reg].append(dataset.loc[dataset["denominazione_regione"] == reg])
        except:
            regions[reg] = dataset.loc[dataset["denominazione_regione"] == reg]
        regions[reg].drop(['stato', 'codice_regione', 'denominazione_regione', 'lat', 'long'], axis = 1, inplace = True)
```

Oridiniamo i dataframes per data

```python
for reg in regions:
    regions[reg].sort_values(by=['data'], inplace = True)
    
regions["Lombardia"]
```

Creiamo adesso un Dataframe raggruppando tutte le informazioni di ogni singola regione per singola data

```python
italy = pd.DataFrame().reindex_like(regions["Abruzzo"])
italy = italy.fillna(0)

for i in range(len(regions["Abruzzo"])):
    for reg in regions:
        italy.iloc[i] += regions[reg].iloc[i]
    
italy
```
