```python
import pandas as pd
import glob
import os
```

```python
states = ["India","Italy","Brazil","Indonesia","South Korea","United Kingdom","United States"]

path_GDP = os.getcwd() + "/data_source/World/GDP/"
path_GDP = path_GDP.replace("/",os.path.sep)
path_Population = os.getcwd() + "/data_source/World/Population/"
path_Population = path_Population.replace("/",os.path.sep)


csv_GDP = glob.glob(path_GDP + "*.csv")
csv_Population = glob.glob(path_Population + "*.csv")

csv_Population

```

```python
GDP_csv = pd.read_csv(csv_GDP[0], index_col = 0)
Population_csv = pd.read_csv(csv_Population[0], index_col = 0)
```

```python
GDP = pd.DataFrame(columns=GDP_csv.columns)

for country in GDP_csv.index:
    if country in states:
        GDP = GDP.append(GDP_csv.loc[country])
        
        
GDP.drop(GDP.columns.difference(['2018']), 1, inplace=True)

GDP
```

```python
Population_csv.set_index('Location', inplace=True)
```

```python
Population_csv.drop(['VarID', 'Variant', 'MidPeriod'], axis=1, inplace = True)
Population_csv
```

```python
path_lockdown = os.getcwd() + "/data_source/world/Lockdown/lockdown.csv"
lockdown = pd.read_csv(path_lockdown, index_col = 0)
```

```python
lockdown.drop(['Applies To', 'Exceptions', 'Implementing City', 'Implementing State/Province', 'Quantity', 'Source', 'Target city', 'Target state', 'Target country', 'Target region'], axis=1, inplace = True)

lockdown = lockdown[(lockdown.Country == "India") | (lockdown.Country == "Italy") | (lockdown.Country == "South Korea") | (lockdown.Country == "United Kingdom") | (lockdown.Country == "United States")]
lockdown = lockdown[lockdown.Keywords.notnull()]
lockdown
```

```python
#get path for os
path = os.getcwd() + "/parsed_data/world/"
path = path.replace("/",os.path.sep)
#save data as CSV 

GDP.to_csv(path + "GDP.csv", index = True, header=True)
Population_csv.to_csv(path + "Population.csv", index = True, header=True)
lockdown.to_csv(path + "Lockdown.csv", index = True, header=True)
```
